```python
#1

import pandas as pd

data = {
    "student_age": [6, 9, 12, 15, 15, 15, 15, 16, 18, 18, 18, 21, 22, 27, 28, 26, 28, 25, 24, 22],
    "hours_of_study": [1, 2, 2, 2, 4, 4, 4, 3, 10, 2, 7, 3, 6, 2, 9, 10, 12, 8, 12, 12],
    "grade": [95, 90, 80, 70, 85, 94, 83, 66, 34, 88, 76, 87, 52, 94, 53, 100, 54, 72, 96, 84]
}

df = pd.DataFrame(data)

df
```

```python
#2
passed = df[df["grade"] >= 55]
print(passed)
```

```python
#3
failed = df[(df["grade"] < 55) & (df["student_age"] > 25)]
print(failed)
```

```python
#4
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.scatter(df["student_age"], df["hours_of_study"], color="blue")
plt.title("Student Age vs Hours of Study")
plt.xlabel("Student Age")
plt.ylabel("Hours of Study")
plt.grid(True)
plt.show()
```

```python
#5
from mpl_toolkits.mplot3d import Axes3D

df_3h = df[df["hours_of_study"] >= 3]

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df_3h["student_age"], df_3h["hours_of_study"], df_3h["grade"], c='green', marker='o')

ax.set_title("3D: Age, Hours of Study, Grade")
ax.set_xlabel("Age")
ax.set_ylabel("Hours of Study")
ax.set_zlabel("Grade")

plt.show()
```

```python
#6
avg_grade = df["grade"].mean()

df["grade_improve"] = df["grade"].apply(lambda g: g if g >= avg_grade else avg_grade)

print(df)
```
